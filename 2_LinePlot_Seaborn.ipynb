### Lineplot in Seaborn

> We are going to work on Lineplot with seaborn on a dataset

#### Importing Libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

##### Creating a basic Lineplot

```python
from turtle import color


x_data=[1,2,3,4,5]
y_data=[1,4,9,16,25]

sns.lineplot(x=x_data,y=y_data,color="red")
plt.show()
```

#### Creating a Lineplot with Dataset

```python
df=sns.load_dataset("iris")
df
```

```python
sns.lineplot(x="sepal_length",y="sepal_width",data=df)
plt.show()
```

#### Parameters of Lineplot

```python
# Hue is used to make graph with respect to Popular data. In this dataset we have three classes.

sns.lineplot(x="sepal_length",y="sepal_width",data=df,hue="species")
plt.show()
```

```python
# Size Parameter

sns.lineplot(x="sepal_length",y="sepal_width",data=df,hue="species", size=90)
plt.show()
```

```python
# Style parameter change the style of class. I am passing the column â€œspeciesâ€
# to style parameter, to group by this column. Each line has a different style.
# Grouping variable that will produce lines with different dashes and/or markers. 

sns.lineplot(x="sepal_length",y="sepal_width",data=df,hue="species",style="species")
plt.show()
```

```python
# palette Parameter is used for coloring the lines

sns.lineplot(x="sepal_length",y="sepal_width",data=df,hue="species",palette="flare")
plt.show()
```

```python
# Marker Parameter is Object determining how to draw the markers for different levels of the style variable.



sns.lineplot(x="sepal_length",y="sepal_width",data=df,hue="species",palette="flare", markers=["o","o","o"])
plt.show()
```

#### To Minimize The Data
