#### Violinplot:

> A violinplot is similar to box plot or whisker plot. It shows the quantitative data across several levels of one categorical values.

> It work on categorical values

> It work on distribution of data

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as  plt
```

```python
data=sns.load_dataset("Tips")
sns.violinplot(x="day",y="total_bill",data=data)
plt.show()
```

#### Parameters

```python
# Width of edge line

sns.violinplot(x="day",y="total_bill",data=data,linewidth=4)
```

```python
# Color of Plot

sns.violinplot(x="day",y="total_bill",data=data,linewidth=2,palette="gnuplot2")
```

```python
# hue of Plot

sns.violinplot(x="day",y="total_bill",data=data,hue="sex")
```

```python
# Saturation of Plot

sns.violinplot(x="day",y="total_bill",data=data,hue="sex",saturation=0.5)
```

```python
# Color of Plot

sns.violinplot(x="day",y="total_bill",data=data,hue="sex",color="r")
```

```python
# splitting of Plot

sns.violinplot(x="day",y="total_bill",data=data,hue="sex",split=True)
```

```python
# Changing the orientation of Plot

sns.violinplot(y="day",x="total_bill",data=data,hue="sex",split=True)
```

```python
# Inner parameter quartz,box,point
# Representation of the datapoints in the violin interior.

sns.violinplot(x="day",y="total_bill",data=data,hue="sex",inner="point")
```

#### Violinplot for Single Column

```python
# Inner parameter quartz,box,point
# Representation of the datapoints in the violin interior.

sns.violinplot(x="total_bill",data=data)
```
