#### Stripplot

>A stripplot is very simple to understand. It is basically a scatter plot that differtiate categories. It is good alternative to a box plot or vilolin plot

>It is used to draw a scatter plot based on the category.

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data=sns.load_dataset("Tips")
sns.stripplot(x="day",y="total_bill",data=data)
```

#### Parameters

```python
#Hue parameter
sns.stripplot(x="day",y="total_bill",data=data,hue="sex")
```

```python
#palette parameter
sns.stripplot(x="day",y="total_bill",data=data,hue="sex",palette="flare")
```

```python
#linewidth parameter
sns.stripplot(x="day",y="total_bill",data=data,hue="sex",linewidth=2)
```

```python
#changing the edgecolor parameter
sns.stripplot(x="day",y="total_bill",data=data,hue="sex",linewidth=2,edgecolor="r")
```

```python
#changing the size of dots parameter
sns.stripplot(x="day",y="total_bill",data=data,hue="sex",size=7)
```
