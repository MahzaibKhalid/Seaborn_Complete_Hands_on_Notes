#### FacetGrid in Seaborn

It is a single graph in multiple one graph

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=sns.load_dataset("Tips")
```

```python
fg=sns.FacetGrid(data)
```

```python
fg.map(plt.scatter,"total_bill","tip")
plt.show()
```

To draw multiple plots in a graph

```python
fg=sns.FacetGrid(data,col="sex",hue="day")
fg.map(plt.scatter,"total_bill","tip")
```

```python
fg=sns.FacetGrid(data,col="day",hue="sex")
fg.map(plt.scatter,"total_bill","tip")
```

```python
fg=sns.FacetGrid(data,col="day",hue="sex")
fg.map(plt.bar,"total_bill","tip")
```

```python
fg=sns.FacetGrid(data,col="day",hue="sex")
fg.map(plt.bar,"total_bill","tip")
```

```python
fg=sns.FacetGrid(data,col="time",hue="day",palette="flare")
fg.map(plt.bar,"total_bill","tip")
```
