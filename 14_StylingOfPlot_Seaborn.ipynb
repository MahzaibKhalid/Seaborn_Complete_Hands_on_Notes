Styling can be done by:

- Seaborn Figure Styles
- Removing axes spines
- Scale and Context

```python
import pandas as pd
import seaborn  as sns
import matplotlib.pyplot as plt
```

```python
data=sns.load_dataset("Tips")
```

```python
sns.set_style("white")
```

```python
sns.countplot(x="sex",data=data)
plt.show() # To remove the adress
```

```python
sns.despine(left= True) # To remove the axis line
sns.countplot(x="sex",data=data)

plt.figure(figsize=(1.5,1.4)) #To change the figure size
```

```python
sns.countplot(x="sex",data=data)

sns.set_context("poster",font_scale=0.1)
```

```python
sns.countplot(x="sex",data=data)

sns.set_context("paper",font_scale=0.2)
```

```python
sns.countplot(x="sex",data=data)

sns.set_context("notebook",font_scale=0.5)
```
