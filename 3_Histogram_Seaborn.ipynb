#### Histogram in Seaborn

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data= sns.load_dataset("iris")
data.columns
```

```python
sns.displot(data["sepal_length"])
plt.show()
```

#### Parameters of Histogram

```python
# bins parameter is used to keep data in fix range. It is Optional parameter contains integer 
# or sequence or strings


sns.displot(data["sepal_length"],bins=[4.5,5.0,5.5,6.0,6.5,7.0,7.5,8.0])
plt.show()
```

```python
# kde Parameter gives a line graph. It stands for kernel density estimator

from weakref import KeyedRef



sns.displot(data["sepal_length"],bins=[4.5,5.0,5.5,6.0,6.5,7.0,7.5,8.0], kde=True)
plt.show()
```

```python
# Color Parameter is used to give color to graph

sns.displot(data["sepal_length"],bins=[4.5,5.0,5.5,6.0,6.5,7.0,7.5,8.0], color="Red")
plt.show()
```

```python
# Color Parameter is used to give color to graph

sns.displot(data["sepal_length"],color="Red",log=True)
plt.show()
```
