#### Scatter Plot in Python

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=sns.load_dataset("penguins")
data.columns
```

```python
sns.scatterplot(x="bill_depth_mm",y="bill_length_mm",data=data)
```

#### Parameter of Scatterplot

```python
sns.scatterplot(x="bill_depth_mm",y="bill_length_mm",data=data,hue="sex")
```

```python
sns.scatterplot(x="bill_depth_mm",y="bill_length_mm",data=data,hue="sex",size="sex")
plt.show()
```

```python
data=sns.load_dataset("penguins").head(20)
sns.scatterplot(x="bill_depth_mm",y="bill_length_mm",data=data,hue="sex",sizes=(90,100),palette="flare")
plt.show()
```

```python
data=sns.load_dataset("penguins").head(20)
sns.scatterplot(x="bill_depth_mm",y="bill_length_mm",data=data,hue="sex",size="sex",palette="flare")
plt.show()
```

```python
data=sns.load_dataset("penguins").head(20)
m={"Male":'*',"Female":'*'}
sns.set_style("whitegrid")

sns.scatterplot(x="bill_depth_mm",y="bill_length_mm",data=data,hue="sex",palette="flare",markers=m)
plt.show()
```
