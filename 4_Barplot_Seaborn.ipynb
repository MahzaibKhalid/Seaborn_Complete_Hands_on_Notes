#### Barplot in Seaborn

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data= sns.load_dataset("penguins")
data
```

```python
sns.barplot(x=data.island,y=data.bill_length_mm)
plt.show()
```

#### Parameters of Barplot

```python
# Hue is used to make graph with respect to Popular data. In this dataset we have three classes.

sns.barplot(x="island",y="bill_length_mm",data=data, hue="sex")
```

```python
# To change order of barplots
order_1=["Dream","Torgersen","Biscoe"]
sns.barplot(x="island",y="bill_length_mm",data=data, hue="sex",order=order_1)
```

```python
# To change hue order of barplots
order_1=["Dream","Torgersen","Biscoe"]
hue_ord=["Female","Male"]
sns.barplot(x="island",y="bill_length_mm",data=data, hue="sex",order=order_1,hue_order=hue_ord)
```

```python
# To use the CI, Confidence interval
sns.barplot(x="island",y="bill_length_mm",data=data, hue="sex",ci=95)
```

```python
# Orientation of Bar plot
# To change the orientation it is mandatory data on both axis should be numerical

sns.barplot(x="bill_depth_mm",y="bill_length_mm",data=data, hue="sex",orient="h")
```

```python
# After Minimizing the Dataset

data=sns.load_dataset("penguins").head(10)
sns.barplot(x="bill_depth_mm",y="bill_length_mm",data=data, hue="sex",orient="h")
```

#### Few More Parameters

```python
# Error Color Increase

data=sns.load_dataset("penguins")
sns.barplot(x="island",y="bill_length_mm",data=data, hue="sex",errcolor="green")
```

```python
# Color Changes

data=sns.load_dataset("penguins")
sns.barplot(x="island",y="bill_length_mm",data=data, hue="sex",color="green")
```

```python
# Palete Changes

data=sns.load_dataset("penguins")
sns.barplot(x="island",y="bill_length_mm",data=data, hue="sex",palette="pink")
```

```python
# Saturation of Bars

data=sns.load_dataset("penguins")
sns.barplot(x="island",y="bill_length_mm",data=data, hue="sex",palette="pink",saturation=10)
plt.show()
```

```python
# Blurring the Bars

data=sns.load_dataset("penguins")
sns.barplot(x="island",y="bill_length_mm",data=data, hue="sex",palette="pink",alpha=0.5)
plt.show()
```

```python
# Capsize of error

data=sns.load_dataset("penguins")
sns.barplot(x="island",y="bill_length_mm",data=data, hue="sex",palette="pink",capsize=0.3)
plt.show()
```
