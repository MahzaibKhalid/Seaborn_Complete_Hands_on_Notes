#### Coutplot:
> The countplot is used to represent the occurrence(counts) of the observation present in the categorical variable. It uses the concept of a bar chart for the visual depiction.


```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=sns.load_dataset("Tips")
data.to_csv("Tips.csv")
```

```python
data1=pd.read_csv("Tips.csv")
```

```python
data1.nunique()
```

```python

sns.countplot(x="day",data=data)
plt.show()
```

#### Parameters

```python

sns.countplot(x="sex",data=data,hue="day")
plt.show()
```

```python
# To change the orientation of Graph
sns.countplot(y="sex",data=data,hue="day")
plt.show()
```

```python

sns.countplot(x="sex",data=data,hue="day",palette="flare")
plt.show()
```

```python

sns.countplot(x="sex",data=data,hue="day",color="r")
plt.show()
```

```python

sns.countplot(x="sex",data=data,hue="day",color="r",saturation=0.4)
plt.show()
```
