#### Pairplot

> A pairplot is a matrix of scatterplot that ets you understand the pairwise relationship between different variables in a dataset

>When you use machine learning algorithm, then it is the need to compare all graphs

> It only works with numerical value only

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data=sns.load_dataset("Tips")
data
```

```python
sns.pairplot(data)
```

#### Parameters

```python
# Hue parameter
sns.pairplot(data,hue="sex")
```

```python
# To use specific columns
sns.pairplot(data,vars=["total_bill","tip"])
```

```python
# To change the order of classes
sns.pairplot(data,hue="sex",hue_order=["Male","Female"])
```

```python
sns.pairplot(data,hue='sex',kind="reg")  # "reg","scatter"."hist"."box"
```

```python
m=["o","*"]
sns.pairplot(data,hue="sex",markers=m)
```
